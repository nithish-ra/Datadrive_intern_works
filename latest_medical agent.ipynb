```python
from fpdf import FPDF

# Initialize the PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", 'B', 16)
pdf.cell(200, 10, txt="CONFIDENTIAL MEDICAL REPORT", ln=1, align='C')
pdf.ln(10)

# Add the mock medical data
pdf.set_font("Arial", size=12)
medical_data = """Patient Name: Alex Mercer
Date of Birth: 1985-04-12
Date of Consultation: March 26, 2026
Attending Physician: Dr. Sarah Jenkins, Pulmonology Dept.

VITAL SIGNS:
- Blood Pressure: 120/78 mmHg
- Heart Rate: 88 bpm
- Temperature: 38.2 C (100.7 F)
- Oxygen Saturation (SpO2): 96% on room air

CHIEF COMPLAINT:
Patient presents with a persistent, dry cough that has lasted for 5 days, accompanied by a mild fever, general fatigue, and slight chest discomfort when coughing. Patient denies any loss of taste or smell.

DIAGNOSIS:
Acute Viral Bronchitis. Lungs are mostly clear to auscultation, with mild wheezing in the upper lobes. No signs of pneumonia at this time.

TREATMENT PLAN & MEDICATIONS:
1. Hydration: Drink at least 2.5 liters of water daily.
2. Rest: Strict voice and physical rest for the next 48 hours.
3. Medication (Fever/Pain): Ibuprofen 400mg every 6 hours as needed for fever or muscle aches.
4. Medication (Breathing): Albuterol sulfate HFA inhaler (90 mcg/actuation) - Take 2 puffs every 4 to 6 hours ONLY if wheezing or shortness of breath occurs.

FOLLOW-UP:
Patient is advised to monitor temperature. If fever exceeds 39.0 C, or if shortness of breath worsens, patient must report to the Emergency Room immediately. Otherwise, follow up in clinic in 7 days if symptoms have not improved."""

# Write the text and save the file
pdf.multi_cell(0, 7, medical_data)
pdf.output("medical_report.pdf")

print("Success! 'medical_report.pdf' has been created.")
```

```python
# ==========================================
# 1. TEXTBOT: RAG PDF Document Setup
# ==========================================
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

print("Loading Medical PDF into Vector Database...")

loader = PyPDFLoader("medical_report.pdf")
docs = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_db = FAISS.from_documents(chunks, embeddings)

print("Medical PDF successfully indexed!")
```

```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, SpeechT5Processor, SpeechT5ForTextToSpeech, SpeechT5HifiGan
import soundfile as sf
from IPython.display import Audio, display

print("Starting Local Models Setup from Saved Disk...")

# --- A. VOICE TO TEXT (Whisper) ---
print("Loading Whisper Model (Local)...")
# ---> NOW POINTING TO LOCAL FOLDER <---
whisper_pipeline = pipeline(
    "automatic-speech-recognition", 
    model="saved_models/whisper", 
    device=0 if torch.cuda.is_available() else -1,
    generate_kwargs={
    "language": "english",
    "task": "transcribe"
})

def convert_voice_to_text(audio_file_path):
    print("Transcribing audio...")
    result = whisper_pipeline(audio_file_path)
    text = result["text"].strip()
    print(f"User Asked: '{text}'")
    return text


from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline
from langchain_community.llms import HuggingFacePipeline
import torch

MODEL_ID = "mistralai/Mistral-7B-Instruct-v0.2"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    quantization_config=bnb_config,
    device_map="auto",
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    do_sample=False,          # deterministic — better for QA
    repetition_penalty=1.15,
    return_full_text=False,   # <--- YOU MUST ADD THIS HERE
)
llm = HuggingFacePipeline(pipeline=pipe)
print("✅ Mistral-7B loaded in 4-bit")

from langchain_community.utilities import WikipediaAPIWrapper
from langchain_community.tools import DuckDuckGoSearchRun

# Initialize external search tools
wikipedia = WikipediaAPIWrapper(top_k_results=1, doc_content_chars_max=500)
web_search = DuckDuckGoSearchRun()

def get_answer_with_rag(user_question):

    print("\nGathering context from multiple sources...")

    # ==========================================
    # SOURCE 1 — LOCAL MEDICAL REPORT
    # ==========================================
    retrieved_docs = vector_db.similarity_search(
        user_question,
        k=3
    )

    local_context = "\n\n".join(
        [doc.page_content for doc in retrieved_docs]
    )

    # ==========================================
    # SOURCE 2 — WIKIPEDIA
    # ==========================================
    try:
        wiki_context = wikipedia.run(user_question)[:300]

        if not wiki_context:
            wiki_context = "No relevant Wikipedia information."

    except Exception:
        wiki_context = "Wikipedia search failed."

    # ==========================================
    # SOURCE 3 — DUCKDUCKGO
    # ==========================================
    try:
        internet_context = web_search.run(user_question)[:300]

        if not internet_context:
            internet_context = "No web search results."

    except Exception:
        internet_context = "Web search failed."

    # ==========================================
    # COMBINED CONTEXT
    # ==========================================
    combined_context = f"""

##############################
PRIMARY SOURCE (HIGHEST PRIORITY)
PATIENT MEDICAL REPORT
##############################

{local_context}


##############################
SECONDARY SOURCE
WIKIPEDIA MEDICAL REFERENCE
##############################

{wiki_context}


##############################
SECONDARY SOURCE
WEB SEARCH REFERENCE
##############################

{internet_context}

"""

    # DEBUGGING
    print("\n=== FINAL COMBINED CONTEXT ===")
    print(combined_context)

    # ==========================================
    # STRICT RAG PROMPT
    # ==========================================
    # ==========================================
# XML STRUCTURED PROMPT
# ==========================================

    # ==========================================
    # XML STRUCTURED PROMPT (FIXED)
    # ==========================================

    prompt = f"""<medical_rag_system>
    
        <system_instruction>
        You are a highly accurate medical AI assistant.
        Read the provided sources and answer the user's question directly and concisely.
    
        STRICT RULES:
        1. ALWAYS prioritize the primary patient report.
        2. NEVER invent patient details, names, or treatment plans.
        3. If information is missing, say: "Information not available."
        4. Keep answers short, natural, and conversational for voice output.
        5. DO NOT output any XML tags in your final answer.
        </system_instruction>
    
        <primary_patient_report>
        {local_context}
        </primary_patient_report>
    
        <wikipedia_medical_reference>
        {wiki_context}
        </wikipedia_medical_reference>
    
        <web_medical_reference>
        {internet_context}
        </web_medical_reference>
    
        <user_question>
        {user_question}
        </user_question>
            
        <response>
"""

    print("\nAnalyzing medical data & public information...")

    # Wrap your XML prompt in an Instruct template
    messages = [
        {"role": "user", "content": prompt}
    ]
    chat_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    # ==========================================
    # GENERATE ANSWER
    # ==========================================
    result = pipe(chat_prompt)

    # Because we set return_full_text=False, this variable now ONLY contains the AI's answer!
    answer = result[0]['generated_text'].strip()
    
    # ==========================================
    # CLEAN LONG OUTPUTS
    # ==========================================
    import re
    answer = re.sub(r"<.*?>", "", answer)
    answer = " ".join(answer.split())
    answer = answer[:450]
    
    print("\nFINAL CLEAN ANSWER:")
    print(answer)
    
    return answer, combined_context

# --- C. TEXT TO VOICE (SpeechT5) ---
print("Loading Text-to-Speech Model (Local)...")
# ---> NOW POINTING TO LOCAL FOLDER <---
tts_processor = SpeechT5Processor.from_pretrained("saved_models/speecht5")
tts_model = SpeechT5ForTextToSpeech.from_pretrained("saved_models/speecht5").to("cuda" if torch.cuda.is_available() else "cpu")

# ---> NOW POINTING TO LOCAL FOLDER <---
vocoder = SpeechT5HifiGan.from_pretrained("saved_models/vocoder").to("cuda" if torch.cuda.is_available() else "cpu")

# ---> LOADING LOCAL VOICE TENSOR <---
speaker_embeddings = torch.load("saved_models/speaker_profile.pt", weights_only=True).to("cuda" if torch.cuda.is_available() else "cpu")

def convert_text_to_voice(text, output_filename="response.wav"):

    print("Generating audio response...")

    inputs = tts_processor(
        text=text,
        return_tensors="pt"
    ).to("cuda" if torch.cuda.is_available() else "cpu")

    speech = tts_model.generate_speech(
        inputs["input_ids"],
        speaker_embeddings,
        vocoder=vocoder
    )

    sf.write(
        output_filename,
        speech.cpu().numpy(),
        samplerate=16000
    )

    print(f"Audio saved as {output_filename}")

    display(Audio(output_filename))

    return output_filename

print("All models loaded successfully from local storage!")
```

```python
# ==========================================
# 5. EXECUTE THE FULL PIPELINE
# ==========================================
def run_textbot_pipeline(audio_input_path):

    print("\n--- STARTING TEXTBOT PIPELINE ---")

    # STEP 1: Voice → Text
    user_prompt = convert_voice_to_text(audio_input_path)

    # STEP 2: RAG + LLM
    final_answer, retrieved_context = get_answer_with_rag(user_prompt)

    # STEP 3: Text → Voice
    convert_text_to_voice(final_answer)


    print("\nPIPELINE COMPLETED SUCCESSFULLY")

print("\nSYSTEM READY!")

# NOTE: Since you are asking about the medical report, upload an audio file asking 
# something like "What is the patient's diagnosis?"
run_textbot_pipeline("/kaggle/input/datasets/nithishra/voice-input-que/voice_input.wav")
```

```python

from fpdf import FPDF
import os
from datetime import datetime

def save_text_to_pdf(user_query, text_content):
    """Saves a timestamped, formatted clinical record to the Kaggle working directory."""
    
    # 1. DYNAMIC NAMING: Use a timestamp so reports are never overwritten!
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_path = f"/kaggle/working/medical_record_{timestamp}.pdf"
    
    # 2. UNICODE SAFETY: FPDF crashes on emojis or weird quotes. This cleans the text.
    safe_query = user_query.encode('latin-1', 'ignore').decode('latin-1')
    safe_answer = text_content.encode('latin-1', 'ignore').decode('latin-1')
    
    pdf = FPDF()
    pdf.add_page()
    
    # --- HEADER ---
    pdf.set_font("Arial", "B", 16)
    pdf.cell(0, 10, "Automated Clinical Analysis", ln=True, align="C")
    pdf.set_font("Arial", "I", 10)
    pdf.cell(0, 10, f"Date Generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}", ln=True, align="C")
    pdf.line(10, 30, 200, 30) # Draws a clean separator line
    pdf.ln(10)
    
    # --- SECTION 1: THE QUERY ---
    pdf.set_font("Arial", "B", 12)
    pdf.cell(0, 8, "1. Voice Query Transcript:", ln=True)
    pdf.set_font("Arial", "I", 11)
    pdf.multi_cell(0, 7, f'"{safe_query}"')
    pdf.ln(5)
    
    # --- SECTION 2: THE AI ANALYSIS ---
    pdf.set_font("Arial", "B", 12)
    pdf.cell(0, 8, "2. RAG Agent Analysis:", ln=True)
    pdf.set_font("Arial", "", 11)
    pdf.multi_cell(0, 7, safe_answer)
    
    # Save to disk
    pdf.output(output_path)
    print(f"📄 Saved Clinical Record -> {output_path}")
```

```python
# ==========================================
# 5. EXECUTE WITH GRADIO WEB UI
# ==========================================
import gradio as gr

# First, make sure we have the gradio package installed
try:
    import gradio as gr
except ImportError:
    print("Installing Gradio...")
    import os
    os.system('pip install -q gradio')
    import gradio as gr

def gradio_pipeline(audio_filepath):

    if not audio_filepath:
        return (
            "No audio detected.",
            "Please record or upload an audio file.",
            None
        )

    print("\n--- NEW INTERACTION ---")

    # STEP 1: Voice → Text
    user_prompt = convert_voice_to_text(audio_filepath)

    # STEP 2: RAG + Reasoning
    final_answer, retrieved_context = get_answer_with_rag(user_prompt)

    # ---------------------------------------------------------
    # NEW STEP: Silently save the final_answer to a PDF!
    # ---------------------------------------------------------
    save_text_to_pdf(user_prompt, final_answer)

    # STEP 3: Text → Voice
    output_audio_path = "gradio_response.wav"

    convert_text_to_voice(
        final_answer,
        output_audio_path
    )

    # RETURN EVERYTHING TO UI (Unchanged)
    return (
        user_prompt,
        final_answer,
        output_audio_path,
    )

# Create the Web UI
iface = gr.Interface(
    fn=gradio_pipeline,
    inputs=gr.Audio(sources=["microphone", "upload"], type="filepath", label="🎙️ Ask the Medical Assistant"),
    outputs=[
    gr.Textbox(label="1️⃣ What the AI heard you say:"),
    gr.Textbox(label="2️⃣ Clinical AI Analysis:"),
    gr.Audio(label="3️⃣ Voice Response")
    ],
    title="🩺 Local Medical RAG Voice Agent",
    description="Ask questions about Alex Mercer's medical report (e.g., 'What is the patient's diagnosis?' or 'What is the treatment plan?'). All processing is done locally on this Kaggle GPU.",
    allow_flagging="never",
    theme=gr.themes.Soft()
)

# Launch the app! 
# share=True creates a public link you can open on your phone to test it there too!
iface.launch(share=True)
```
